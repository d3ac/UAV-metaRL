```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

index_max = 0

data1 = pd.read_csv('data/reward_lstm.csv').values[:,1]
data2 = pd.read_csv('data/reward_dqn.csv').values[:,1]
index_max = min(len(data1), len(data2))
data1 = data1[:index_max]
data2 = data2[:index_max]

x = np.arange(len(data1))

def moving_average(data, window_size):
    processed_data = []
    for i in range(len(data)):
        if i < window_size:
            processed_data.append(np.mean(data[:i+1]))
        else:
            processed_data.append(np.mean(data[i-window_size:i]))
    return processed_data

data1_moving = moving_average(data1, 20)
data2_moving = moving_average(data2, 20)

plt.plot(x, data1, color='#e75840', alpha=0.3)
plt.plot(x, data1_moving, color='#e75840', label='lstm')

plt.plot(x, data2, color='#4e79a7', alpha=0.3)
plt.plot(x, data2_moving, color='#4e79a7', label='dqn')

plt.legend()
plt.grid(True)
plt.show()
```
